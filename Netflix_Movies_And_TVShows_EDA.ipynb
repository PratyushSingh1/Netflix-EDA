# **Checking of Null values**

```python
df.isna().sum()
```

# **Here we have replaced No values with na**

```python
df["director"].fillna("No Director", inplace = True)
df["cast"].fillna("No Cast", inplace = True)
df["country"].fillna("Country Unavailable", inplace = True)
```

# **We removed all the rows with numbers which will not affect our EDA**

```python
df.dropna(subset = ['date_added', 'rating', 'duration'], inplace = True)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()   
```

```python
df['type'].value_counts()
```

```python
sns.countplot(x='type', data=df , palette='magma')
plt.title('Movies VS TV Shows')
plt.xlabel('Show Types')
plt.ylabel('counts (Total Number)')
plt.show()
```

*Here we can see that there are a lot of movies on Netflix as compared to TV Shows*

---


```python
import plotly.express as px
x=df["type"].value_counts().reset_index()
px.pie(x,values="type",names="index")
```

*Here we can clearly see Movies are 69.7% which brings a lot of difference as compared to TV Shows* 

---



```python
top_10_countries= df.country.value_counts().head(10)
top_10_countries = pd.DataFrame(top_10_countries)
top_10_countries
```

```python
top_10 = df['country'].value_counts()[:10]
top_10.plot(kind='bar',figsize=(10,8))
plt.title('Countries which produces more Movies and TV Shows')
plt.xlabel('Top 10 Countries')
plt.ylabel('Total Counts')
plt.show()
```

*we can see that United states tops the list in terms of Movies and TV shows Production and views followed by India and united Kingdom, well thats a huge difference United States takes a huge lead* 


---






```python
ten_yrs=df[["type","release_year"]]
ten_yrs=ten_yrs[ten_yrs["release_year"] >=2012]
ten_yrs
```

```python
ten_yrs.groupby("release_year")["type"].value_counts()
```

```python
plt.figure(figsize=(18,9))
sns.set(style="dark")
ax = sns.countplot(y="release_year", data=df, palette="magma", order=df['release_year'].value_counts().index[0:15])
```

*Here we can see Most of the releases took place in 2018 and than 2017 and 2019 almost showed no difference. Here we can predict that since 2006 Netflix is releasing of Movies and TV Shows had increased and hence Netflix did a good Profit*

---



```python
plt.figure(figsize=(10,8))
sns.countplot(x="release_year", data=ten_yrs, hue="type", palette="flare")
```

*It can be seen clearly that Movies and TV Shows Production is increasing year by year but after 2019 Movies got huge down fall and after 2020 both Movies and TV Shows droped the production but there is a great increase in Movie Production as compared to TV Shows But its not Like that TV Shows Product didn't do well but they did. So here we can say that after 2019 production and release on Netflix falled down!*

This are the films which can be seen by small Childrens*
